Image Labelling

```python
# Import necessary libraries
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Set the path to the dataset folder
data_path = "Sample"

# Create an empty list to store the images and labels
images = []
labels = []

# Loop through the dataset folder and read the images
for folder in os.listdir(data_path):
    print(folder)
    folder_path = os.path.join(data_path, folder)
    print(folder_path)
    for file in os.listdir(folder_path):
        print(file)
        file_path = os.path.join(folder_path, file)
        image = cv2.imread(file_path)
        image = cv2.resize(image,(100, 100))
        images.append(image)
        labels.append(folder)

# Convert the images and labels to NumPy arrays
images = np.array(images)
labels = np.array(labels)
print("labeled")
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
print("test train splitting")

# Flatten the images and normalize the pixel values
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255
print("test train splitted")
```

Model Training and Testing

```python
# Train a Support Vector Machine (SVM) model
model = SVC(kernel='linear', C=1, random_state=42)
model.fit(X_train, y_train)

print("model created")

# Predict the labels of the testing set and calculate the accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Deploy the model in the smart dustbin system
# Capture an image of the waste material and pass it through the model to determine whether it is plastic or not
```
